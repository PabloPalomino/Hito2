3.1.- Grafico de barras

```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("https://raw.githubusercontent.com/PabloPalomino/Hito2/main/Indian_Cities_Database.csv")

top=df[0:10]

plt.pie(top["Lat"],labels=top["City"],
        startangle=320,
        radius = 1.2, autopct = '%1.2f%%')


plt.title('Latitud')
plt.show()
```

3.2.- Grafico de barras

```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("https://raw.githubusercontent.com/PabloPalomino/Hito2/main/Indian_Cities_Database.csv")

top=df[0:10]

plt = top.plot.bar (y = ["Lat"], x = "City")
```

3.3.- Grafico combinado

```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("https://raw.githubusercontent.com/PabloPalomino/Hito2/main/Indian_Cities_Database.csv")

top=df[0:10]

ax = top['Lat'].plot(kind='bar',grid=True)
ax2 = ax.twinx()
ax2.plot(top['Long'], linestyle='-', linewidth=2.0,color='red')
plt.show();
```
